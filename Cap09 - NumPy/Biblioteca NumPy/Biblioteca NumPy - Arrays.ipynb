```python
import numpy as np
```

# Criando Arrays NumPy

```python
# Array criado a partir de uma lista
arr1 = np.array([10, 21, 32, 43, 48, 15, 76, 57, 89])
```

```python
print(arr1)
```

```python
type(arr1)
```

```python
arr1.shape
```

```python
# Imprimindo um elemento
arr1[4]
```

```python
arr1[1:4]
```

```python
# Criar uma lista de índices
indices = [1, 2, 5, 6]
arr1[indices]
```

```python
# Cria um máscara booleana para os elementos
mask = (arr1 % 2 == 0)
arr1[mask]
```

```python
# Aterando um elemento do array
arr1[0] = 100
print(arr1)
```

```python
# Não é possível incluir elemento de outro tripo
try:
  arr1[0] = 'novo elemento'
except:
  print('Operação não permitida')
```

# Funções NumPy

```python
# Funçao array() cria um array Numpy
arr2 = np.array([1, 2, 3, 4, 5])
print(arr2)
```

```python
# Métodos do array Numpy
arr2.cumsum()
```

```python
arr2.cumprod()
```

```python
# Funçao cria um array NumPy contendo uma progressão aritimética aparti de um intervalo - start, stop, astep
arr3 = np.arange(0, 50, 3)
print(arr3)
```

```python
# Criando array com elementos zero
arr4 = np.zeros(10)
print(arr4)
```

```python
# Retorna matriz identidade
arr5 = np.eye(3)
print(arr5)
```

```python
# Matriz diagonal
arr6 = np.diag(np.array([1, 2, 3, 4]))
print(arr6)
```

```python
# Conjutno de valores numérico
print(np.linspace(0, 10, 15))
print(np.logspace(0, 5, 10))
```

# Operações com Arrays

```python
# Somando arrays
arr19 = np.array([[1, 2], [3, 4]])
arr20 = np.array([[5, 6], [0, 7]])

print(arr19)
print(arr20)
```

```python
# Soma de Arrays
np.add(arr19, arr20)
```

```python
# Multiplicação de Matrizes
np.dot(arr19, arr20)
```

```python
# Multiplicação de Matrizes
arr21 = arr19 @ arr20
print(arr21)
```
