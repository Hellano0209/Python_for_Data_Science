### Exercício 1 - Imprima na tela os números de 1 a 10. Use uma lista para armazenar os números.

```python
lista = [] # criando uma lista vazia

# Loop onde a variável 'i' vai percorrer os valores 1 a 10
# e a cada interação o valor de 'i' é gravado na variável 'lista'
for i in range(1, 11):
  lista.append(i)

print(lista)
```
