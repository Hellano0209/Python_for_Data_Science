# Gráfico de Linhas

```python
from pylab import *
```

```python
# Gráfico de linha

# Dados
x = linspace(0, 5, 10)
y = x**2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0, 0, 0.8, 0.8])

# Cria o plot
axes.plot(x, y, 'r')

# Labels e título
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

```python
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de Linha')

plt.show()
```

```python
# Gráfico de linha com 2 figuras

# Dados
x = linspace(0, 5, 10)
y = x**2

# Criar a figura
fig = plt.figure()

# Cria os eixos
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eixo da figura principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eixo da figura secundária

# Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# Figura secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Figura Secundária')
```

```python
# Gráficos de linha em Paralelo

# Dados
x = linspace(0, 5, 10)
y = x**2

# Divide a áre de plotagem em dois subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot
for ax in axes:
  ax.plot(x, y, 'r')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('Título')

# Ajustar o layout
fig.tight_layout()
```

```python
# Gráfico de linha com diferentes escalas

# Dados
x = linspace(0, 5, 10)
y = x**2

# Criar os subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 4))

# Cria o plot 1
axes[0].plot(x, y, x, exp(x))
axes[0].set_title('Escala Padrão')

# Cria o plot 2
axes[1].plot(x, y, x, exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)')
```

```python
# Grid

# Dados
x = linspace(0, 5, 10)
y = x**2

# Criar os subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 3))

# Grid padrão
axes[0].plot(x, y, x, exp(x), lw = 2)
axes[0].grid(True)
axes[0].set_title('Escala Padrão')

# Cria o plot 2
axes[1].plot(x, y, x, exp(x))
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)')
```

# Histograma

```python
# Dados
n = np.random.randn(10000)

# Cria os subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 4))

# Plot 1
axes[0].hist(n)
axes[0].set_title('Histograma Padrão')
axes[0].set_xlim((min(n), max(n)))

# Plot 2
axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title('Histograma Cumulativo')
axes[1].set_xlim((min(n), max(n)))
```
