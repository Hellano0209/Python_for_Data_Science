# Statistical Data Visualization com Seaborn

```python
import random
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import seaborn as sea
import warnings

warnings.filterwarnings('ignore')
```

```python
# Carregando um dos datasets que vem com o Seaborn
dados = sea.load_dataset('tips')
```

```python
dados.head()
```

```python
# Método joinplot cria plot de 2 variáveis com gráficos bivariados e univariados
sea.jointplot(data = dados, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
# Método lmplot() cria plot com dados e modelos de regressão
sea.lmplot(data = dados, x = 'total_bill', y = 'tip', col = 'smoker')
```

```python
# Construir um dataframe com pandas
df = pd.DataFrame()
```

```python
# Alimentando o Dataframe com valores aleatórios
df['idade'] = random.sample(range(20, 100), 30)
df['peso'] = random.sample(range(55, 150), 30)
```

```python
df.shape
```

df.head()

```python
# lmplot
sea.lmplot(data = df, x = 'idade', y = 'peso', fit_reg = True)
```

```python
# kdeplot
sea.kdeplot(df.idade)
```

```python
# kdeplot
sea.kdeplot(df.peso)
```

```python
# distplot
sea.distplot(df.peso)
```

```python
# Histograma
plt.hist(df.idade, alpha = 0.3)
sea.rugplot(df.idade)
```

```python
# Boxplot
sea.boxplot(df.idade, color = 'm')
```

```python
# boxplot
sea.boxplot(df.peso)
```

```python
# Violin plot
sea.violinplot(df.idade, color = 'g')
```

```python
#violin plot
sea.violinplot(df.peso, color = 'cyan')
```

```python
# Clustermap
sea.clustermap(df)
```

# Usando Matplotlib, Seaborn, NumPy e Pandas na Criação de Gráficos Estatísticos

```python
# Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variáveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(10, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# DataFrame
dados = pd.DataFrame({'altura':altura, 'peso':peso, 'flag_fumante':flag_fumante})

# Cria os dados para a variável flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True:'Fumante', False:'Não fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Style
sea.set(style = 'ticks')

# lmplot
sea.lmplot(
           data = dados,
           x = 'altura',
           y = 'peso',
           hue = 'flag_fumante', # preencher o conteúde de acordo com uma variável
           palette = ['tab:blue', 'tab:orange'],
           height = 7
           )

# Labels e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação entre Altura e Peso de Fumantes e Não Fumantes')

# Remove as bordas
sea.despine()

plt.show()
```
